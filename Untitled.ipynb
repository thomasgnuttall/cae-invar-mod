```python
import numpy as np
import matplotlib.pyplot as plt
brova = np.load('output/akkarai/brova_barama.mp3.npy')
```

```python
plt.figure(figsize=(100,100))
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)

ax.xaxis.set_ticks(np.arange(0, len(brova), 50))
ax.yaxis.set_ticks(np.arange(0, len(brova), 50))
ax.tick_params(labelrotation=45)

ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.grid()
im = ax.imshow(brova, cmap=plt.get_cmap('hot'))
```

```python
import librosa
audio_path = "./audio/brova_barama.mp3"
audio_loaded, sr = librosa.load(audio_path, sr=22050)
```

```python
window_size = len(audio_loaded)/len(brova)
```

```python
start = [2740, 2650, 1750, 1320]
for s in start:
    point = s*window_size
    seconds = point/22050
    print(seconds)
    minutes = seconds//60
    seconds = seconds%60
    print(f'{int(minutes)}:{round(seconds,2)}')
```
